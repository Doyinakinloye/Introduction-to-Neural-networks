### Creating random tensors
Random tensors are tensors of some arbitrary size which contain random numbers.

```python
# Create two random (but the same) tensor
random_1 = tf.random.Generator.from_seed(42) #set seed reproducibility
random_1 = random_1.normal(shape=(3, 2))
random_1
```

### Shuffling the order of elements in a tensor

```python
# Shuffle a tensor (valuable for when you want to shuffle your data so the inherent order would'nt affect learning)
not_shuffled = tf.constant([[10, 7],
                             [3, 4],
                            [2, 5]])
# Shuffle our non-shuffled tensor
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled, 42)
```

**Aggregating Tensors**

Condensing them from multiple values dow to a smaller amount of values

Forms of Aggregation:
* Get the minimum
* Get the maximum
* Get the mean
* Get the sum


```python
 Tensor = tf.constant(np.random.randint(0, 100, size=50))
```

```python
# Get the mean
Mean = tf.reduce_mean(Tensor)
print(Mean.numpy())
```

```python
# Get the sum
Sum = tf.reduce_sum(Tensor)
print(Sum.numpy())
```

```python
# Get the Maximum
Max = tf.reduce_max(Tensor)
print(Max.numpy())
```

```python
# Get the Minimum
Min = tf.reduce_min(Tensor)
print(Min.numpy())
```

```python
# Find the variance of our tensor
Var = tf.math.reduce_variance(tf.cast(Tensor, dtype=tf.float32))
print(Var.numpy())
```

```python
# Find the standard deviation of tensors
Std = tf.math.reduce_std(tf.cast(Tensor, dtype=tf.float32))
print(Std.numpy())
```

### Find the Positional minimum and maximum

This entails how to find the position(Index) of maximum or minimum tensors


```python
# Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
pos_max = tf.argmax(F)
print(pos_max.numpy())
```

```python
assert F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positionmal minimum
pos_min = tf.argmin(F)
print(pos_min.numpy())
```
